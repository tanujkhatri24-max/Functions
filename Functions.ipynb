6. Why are lists mutable, but tuples are immutable?
   - Lists are mutable because their elements can be changed, added, or  
     removed after the list is created.
     For example, we can update a value in a list or insert a new one.

    - Tuples are immutable, which means once a tuple is created, we cannot    change its values. This makes tuples faster and safer to use when we don’t want the data to be modified.

8. What are logical operators in Python?

10. What is the difference between implicit and explicit type casting?

11. What is the purpose of conditional statements in Python?  
     -  The purpose of conditional statements in Python is to make decisions
        in a program.
      - They allow the program to run different code depending on whether a
        condition is True or False.

      For example, using **if, elif, else**, we can check conditions and control the flow of the program.

Example:

```python
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```
12. How does the elif statement work in Python?
    -  The `elif` statement in Python is used when we want to check **more than one condition**.
     -  It means **“else if”**. If the first `if` condition is False, then
       Python checks the `elif` condition.
       If that is also False, it moves to the next `elif` or the `else` part.

Example:

```python
x = 20

if x < 10:
    print("x is less than 10")
elif x < 30:
    print("x is less than 30")
else:
    print("x is 30 or more")
```
13. What is the difference between FOR and WHILE loop?
     -   Both **for** and **while** loops are used to repeat a block of code,
      but they work in different ways:

 * **For loop:**

      * Used when we know how many times we want to run the loop.
      * Works with sequences like lists, strings, or a range of numbers.
      * Example:

     ```python
      for i in range(5):
      print(i)   # prints 0 to 4
      ```

* **While loop:**

  * Used when we do not know the exact number of repetitions.
  * Runs as long as the condition is True.
  * Example:

    ```python
      x = 0
      while x < 5:
      print(x)
     x += 1
      ```

9. Write a pyhton program to convert user input to string to integer, float, and booolean types.?

```python
# Program to convert user input into different types

user_input = input("Enter the value: ")

# convert to string
str_val = str(user_input)
print("String:", str_val)

# convert to integer
try:
    int_val = int(user_input)
    print("Integer:", int_val)
except:
    print("Cannot convert to Integer")

# convert to float
try:
    float_val = float(user_input)
    print("Float:", float_val)
except:
    print("Cannot convert to Float")

# convert to boolean
bool_val = bool(user_input)
print("Boolean:", bool_val)
```

10. Write code to demonstrate type casting with list element.?

```python
# Program to show type casting with list elements

numbers = ["10", "20", "30"]

print("Original list:", numbers)

# to integer
print("As Integers:", [int(x) for x in numbers])

# to float
print("As Floats:", [float(x) for x in numbers])

# to boolean
print("As Booleans:", [bool(x) for x in numbers])
```

**Output**:

```
Original list: ['10', '20', '30']
As Integers: [10, 20, 30]
As Floats: [10.0, 20.0, 30.0]
As Booleans: [True, True, True]
```

14. Write a program to reverse a string using a while loop.?

```python
# Program to reverse a string using while loop

text = input("Enter a string: ")

reversed_text
i = len(text) - 1

while i >= 0:
    reversed_text = reversed_text + text[i]
    i = i - 1

print("Reversed string is:", reversed_text)
```

15.  Write a pyhton program  to calculate the factorial of a number provided by the user using a while loop.?

```python
# Program to calculate factorial using while loop

# take input from user
num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```

**Output:**

```
Enter a number: 5
Factorial of 5 is: 120
```
